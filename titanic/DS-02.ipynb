```python
def sign(n):
  if n > 0:
    return '양수'
  elif n < 0:
    return '음수'
  else:
    return '0'

print(sign(-1))
print(sign(2))
print(sign(0))
```

```python
def factorial(n):
  result = 1

  for i in range(1, n+1):
    result = result * i
    
  return result
    
print(factorial(4))
```

```python
def factorial(n):
  if n == 1:
    return 1
  else:
    return n * factorial(n-1)
    
print(factorial(4))
```

```python
def cycle(n):
  result = [n]

  while n != 1:
    if n % 2 == 0:
      n = n / 2
    else:
      n = n * 3 + 1
    
    result.append(n)
    
  return result

print(cycle(3))
```

```python
def cycle(n):
  if n == 1:
    return [n]
  
  if n % 2 == 0:
    return [n] + cycle(n/2)
  else:
    return [n] + cycle(3 * n + 1)

cycle(3)
```

```python
def game369(maximum):
  result = []

  for n in range(1, maximum + 1):
    number = str(n)
    
    if '3' in number or '6' in number or '9' in number:
        result.append("clap!")
    else:
        result.append(number)
        
  return result

game369(40)
```
